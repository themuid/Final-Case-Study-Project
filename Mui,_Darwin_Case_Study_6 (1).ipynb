<font size="5">1st Question:</font>

Now, we will aim to use (EDA) to answer our 1st question of the Case Study.

"What can we say about the top 10 video games sold in terms of JP Sales?"

```python
df_videogame_sales = df[['Rank', 'Name', 'Platform', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].head(20)
df_videogame_sales.sort_values('JP_Sales', ascending= False, inplace=True)
df_top_videogame_sales = df_videogame_sales.rename(columns={'Global_Sales':'Worldwide_Sales'})

df_top_videogame_sales
```

```python
fig = px.bar(
    df_top_videogame_sales, 
    x='Name', 
    y='JP_Sales',
    color = 'Name',
    title='Top 20 Video Game Sales In Japan',
    text = 'JP_Sales',
    width = 900,
    height = 900
)
fig.show()
```

<font size="5">Takeaways:</font>

By analyzing our graph, we can see that the top video game sold in terms of Japan Sales are Pokemon games, making sense due to the fandom/popularity/reach it has over there. We can also see that near the bottom, there are more violent games such as Grand Theft Auto that might not be bought/played as much in Japan.

<font size="5">2nd Question:</font>

Now, we will aim to use (EDA) to answer our 2nd question of the Case Study.

"How many top video games in terms of sales were released in the year 2003?"

<font size="5">Key Reminder:</font>

Before we do our initial code, in order to compare "Year" and "Name" variables, they both have to be the same data type. From the df.info() ran earlier in this projectm we know that "Year" is "float64", while "Name is an "object". Thus, we just have to run some code to make it so that they're both objects.

```python
df['Year'] = df['Year'].astype(str)
```

Now, we can run code to compare the two, since they're the same data type.

```python
df_videogame_sales_by_year = df.groupby('Year', as_index = False).agg({
    'Name': 'count',
})

df_videogame_sales_by_year.rename(columns={'Name':'Num_VideoGames'}, inplace = True)
df_videogame_sales_by_year.sort_values(['Num_VideoGames'], ascending = False, inplace= True)

df_videogame_sales_in_2000s = df_videogame_sales_by_year[(df_videogame_sales_by_year['Year']>= '2000.0')]

df_videogame_sales_in_2000s
```

```python
fig = px.treemap(
    df_videogame_sales_in_2000s,
    path= ['Year'],
    title='Number of Top Selling Video Games Released in 2000s Breakdown',
    values='Num_VideoGames',
    height=1000
)

fig.show()
```

<font size="5">Takeaways:</font>

By analyzing our visualization, we can see that in the year 2003, there were 775 games that all sold over 100,000 copies. We also see the amount of top selling video games sold in other years of the 2000s.

<font size="5">3rd Question:</font>

Now, we will aim to use (EDA) to answer our 3rd question of the Case Study.

"What can we say about the distribution of global_sales for these video games?"

```python
fig = px.box(
    df, 
    x='Global_Sales',
    orientation='h',
    title='Video Game Global Sales Distribution (Horizontal)',
    width=16000,
    template = 'ggplot2'
)
fig.show()
```

<font size="5">Takeaways:</font>

By analyzing our visualization, we can see that in our data set, for Global_Sales, the min copies sold for any video game is 0.01, the max copies sold for any video game is 82.74, and the median copies sold for any video game is 0.17. As a reminder, these numbers are all in terms of millions. 

<font size="5">4th Question:</font>

Now, we will aim to use (EDA) to answer our 4th question of the Case Study.

"What is the year in which the least amount of top selling video games released in a year since 2000, and what is that number?"

```python
fig = px.histogram(
    df, 
    x='Year',
    title = 'Number of Top Selling Video Games By Year',
    nbins = 80,
    height = 3000,
    template='plotly_dark'
)
fig.show()
```

<font size="5">Takeaways:</font>

By analyzing our visualization, we can see that since the year 2000, the year with the least amount of video games selling over 100,000 copies is the year 2020, with only 1 game. This makes sense, as since the game has been released so recently, it hasn't had enough time to sell as many copies as games released earier in this dataset.

<font size="5">5th Question:</font>

Now, we will aim to use (EDA) to answer our 5th question of the Case Study.

"What can we infer about a heatmap comparing the "Year" and "Genre" variables for Video Game Sales?"

```python
fig = px.density_heatmap(
    df, 
    x='Year', 
    y='Genre',
    title='Video Game Global Sales Distribution (Horizontal)'
    )
fig.show()
```

<font size="5">Takeaways:</font>

By analyzing our visualization, we can see that in the period between 2008-2009, the genre which had the most video games selling over 100,000 copies is "Action", with 493 games. We can also see that the most successful genres for video game sales are Adventure, Action, Shooter, and Sports, as they glow brightest on this heatmap. meaning the count is higher.

<font size="5">6th Question:</font>

Now, we will aim to use (EDA) to answer our 6 question of the Case Study.

"What can a heatmap made with our dataset tell us about Video Game Sales?"

```python
fig = px.sunburst(
    df,
    path = ['Platform', 'Genre'],
    values = 'Global_Sales',
    height = 1000,
    title = 'Video Game Sales by Platform and Genre'
)
fig.show()
```

<font size="5">Takeaways:</font>

By analyzing our visualization, we can see that the Platform/Console with the most video game sales is the PS2 with 1,255.64 in Global Sales. Remember that these numbers are expressed in millions of copies. This analysis makes sense as the PS2 is one of the oldest consoles out, with many fun games to play on, and creating a lot of nostalgia for the older generation. It makes sense that it has the most video game sales.
